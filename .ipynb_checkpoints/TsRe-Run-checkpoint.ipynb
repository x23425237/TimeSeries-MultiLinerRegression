```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from pmdarima import arima as pmd
from statsmodels.tsa.arima.model import ARIMA
```

```python
df = pd.read_csv('../Stats_CA/Data/ts7.csv')
df.info()
del df['Unnamed: 0']

df.plot()
```

```python
# split a DataFrame (df) into training and testing sets based on the index
testtrain = (df.index < len(df)-30)
df_train = df[testtrain].copy()
df_test = df[~testtrain].copy()
```

```python
# Check for stationarity of time series
acf_check = plot_acf(df_train)

pacf_check = plot_pacf(df_train)
```

```python
# ADF test

def test_adf(data):
    moving_avg=data.rolling(12).mean()
    moving_std=data.rolling(12).std()
    plt.figure(figsize=(20,10))
    plt.plot(data,color='blue',label='Orignial')
    plt.plot(moving_avg,color='red',label='Rolling Mean')
    plt.plot(moving_std,color='black',label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show(block=False)
    print("Results of dickey fuller test")
    adft=adfuller(data,autolag='AIC')
    output=pd.Series(adft[0:4],index=['Test stats','p-value','No of lags used','Number of observations used'])
    for key,value in adft[4].items():
        output['Critical valu (%s)'%key]=value
    print(output)

test_adf(df['x'])  
```

```python
# Kwiatkowski-Phillips-Schmidt-Shin (“KPSS”) test
from statsmodels.tsa.stattools import kpss

def kpss_test(timeseries):
    print("Results of KPSS Test:")
    kpsstest = kpss(timeseries, regression="c", nlags="auto")
    kpss_output = pd.Series(
        kpsstest[0:3], index=["Test Statistic", "p-value", "Lags Used"]
    )
    for key, value in kpsstest[3].items():  # Corrected this line
        kpss_output["Critical Value (%s)" % key] = value
    print(kpss_output)

```

```python
kpss_test(df["x"])
```

```python
# Apply Differencing to make the time series stationary 

df_train_diff = df_train.diff().dropna()
df_train_diff.plot()
# Add title
plt.title('Differenced time series plot')
```

```python
# perform ACF and PACF checks on differenced data 
# note acf and pacf are same 
acf_diff = plot_acf(df_train_diff,lags=25)

pacf_diff = plot_pacf(df_train_diff,lags=25)
```

```python
# perform ADF check. Rolling mean, SD and Differenced time series.
print ( test_adf(df_train_diff['x']) )
```

```python
# Determine ARIMA models parameters p, q, Fit the ARIMA model

model = ARIMA(df_train, order=(0,1,0))
model_fit1 = model.fit()
print(model_fit1.summary())
```

```python
model_fit1.plot_diagnostics(figsize=(10, 8))
```

```python
model_fit1.plot_diagnostics(figsize=(10, 8))
```

```python
# Forecast on test data 
forecast_test = model_fit1.forecast(len(df_test))

df['forecast_manual'] = [None]*len(df_train) + list(forecast_test)
```

```python


# Generate forecast with confidence intervals using get_forecast()
forecast_obj = model_fit1.get_forecast(steps=len(df_test))  # Generate forecast object
forecast_test = forecast_obj.predicted_mean  # Extract forecasted values
conf_int = forecast_obj.conf_int(alpha=0.05)  # 95% confidence intervals

# Ensure forecast indices align with df_test
forecast_test.index = df_test.index
conf_int.index = df_test.index

# Add the forecasted values to the original DataFrame
df['forecast_manual'] = None  # Initialize the column
df.loc[df_test.index, 'forecast_manual'] = forecast_test  # Assign forecasted values

# Plotting the original data (train + test) and the forecasted values
plt.figure(figsize=(10, 6))

# Plot manual forecast
plt.plot(df['forecast_manual'], label='Manual Forecast', color='red')

# Plot actual values
df['actual_values'] = df['x']  # Assuming 'x' contains the actual target values
plt.plot(df['actual_values'], label='Actual Values', color='green')

# Plot confidence intervals
plt.fill_between(
    conf_int.index,
    conf_int.iloc[:, 0],  # Lower bound
    conf_int.iloc[:, 1],  # Upper bound
    color='pink',
    alpha=0.3,
    label='95% Confidence Interval'
)

# Add a legend, labels, and a title
plt.legend()
plt.title('ARIMA Forecast vs Actual with Confidence Interval')
plt.xlabel('Date')
plt.ylabel('Value')

# Show the plot
plt.show()
```

```python
# Model Predictions 
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(df_test, forecast_test)
mape = mean_absolute_percentage_error(df_test, forecast_test)
rmse = np.sqrt(mean_squared_error(df_test, forecast_test))

print(f'mae - manual: {mae}')
print(f'mape - manual: {mape}')
print(f'rmse - manual: {rmse}')
```
