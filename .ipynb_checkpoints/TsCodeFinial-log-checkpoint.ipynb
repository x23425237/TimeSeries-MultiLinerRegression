```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from pmdarima import arima as pmd
from statsmodels.tsa.arima.model import ARIMA
```

```python
df = pd.read_csv('../Stats_CA/Data/ts7.csv')
df.info()
del df['Unnamed: 0']

df.plot()
```

```python
from statsmodels.tsa.api import SimpleExpSmoothing
model = SimpleExpSmoothing(df)
model_single_fit = model.fit()
```

```python
forecast_single = model_single_fit.forecast(6)
2
print(forecast_single)
```

```python
forecast_single40 = model_single_fit.forecast(40)
```

```python
plt.plot(df, label='Original Data')
plt.plot(model_single_fit.fittedvalues, label='Fitted Values')
plt.plot(forecast_single40, label='Forecast')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.title('Single Exponential Smoothing')
plt.legend()
plt.show()
```

```python
def test_adf(data):
    moving_avg=data.rolling(12).mean()
    moving_std=data.rolling(12).std()
    plt.figure(figsize=(20,10))
    plt.plot(data,color='blue',label='Orignial')
    plt.plot(moving_avg,color='red',label='Rolling Mean')
    plt.plot(moving_std,color='black',label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show(block=False)
    print("Results of dickey fuller test")
    adft=adfuller(data,autolag='AIC')
    output=pd.Series(adft[0:4],index=['Test stats','p-value','No of lags used','Number of observations used'])
    for key,value in adft[4].items():
        output['Critical valu (%s)'%key]=value
    print(output)

    
                     
  
```

```python
 test_adf(df['x']) 
```

```python
# log of the x column 
from pylab import rcParams
rcParams['figure.figsize']= 20,10
data_log=np.log(df)
moving_avg_log=data_log.rolling(12).mean()
std_dev_log=data_log.rolling(12).std()
#plt.legend(loc='best')
plt.title('Moving average ')
plt.plot(std_dev_log,color='black',label='Standard Deviation')
plt.plot(moving_avg_log,color='red',label='Mean')
plt.legend()
plt.show()

```

```python
 test_adf(data_log['x']) 
```
