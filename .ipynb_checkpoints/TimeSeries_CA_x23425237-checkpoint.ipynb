# ADF-Augmented-Dickey-Fuller-Test-and-KSS-Kwiatkowski-Phillips-Schmidt-Shin-test-for-stationarity

```python
# perform ADF and KSS test to check for stationality in the data 
def test_adf(data):
    moving_avg=data.rolling(12).mean()
    moving_std=data.rolling(12).std()
    plt.figure(figsize=(20,10))
    plt.plot(data,color='blue',label='Orignial')
    plt.plot(moving_avg,color='red',label='Rolling Mean')
    plt.plot(moving_std,color='black',label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show(block=False)
    print("Results of dickey fuller test")
    adft=adfuller(data,autolag='AIC')
    output=pd.Series(adft[0:4],index=['Test stats','p-value','No of lags used','Number of observations used'])
    for key,value in adft[4].items():
        output['Critical valu (%s)'%key]=value
    print(output)
```

```python
 # Adf test 
test_adf(df['x']) 
```

```python
# KPSS test
t_stat, p_value, _, critical_values = kpss(df['x'].values, nlags='auto')

print(f'ADF Statistic: {t_stat:.2f}')
for key, value in critical_values.items():
     print('Critial Values:')
     print(f'   {key}, {value:.2f}')

print(f'\np-value: {p_value:.2f}')
print("Stationary") if p_value > 0.05 else print("Non-Stationary")
```

```python
# 95% confidence interval for model 1 

# Generate forecast with confidence intervals using get_forecast()
forecast_obj = model_fit1.get_forecast(steps=len(df_test))  # Generate forecast object
forecast_test = forecast_obj.predicted_mean  # Extract forecasted values
conf_int = forecast_obj.conf_int(alpha=0.05)  # 95% confidence intervals

# Ensure forecast indices align with df_test
forecast_test.index = df_test.index
conf_int.index = df_test.index

# Add the forecasted values to the original DataFrame
df['forecast_manual'] = None  # Initialize the column
df.loc[df_test.index, 'forecast_manual'] = forecast_test  # Assign forecasted values

# Plotting the original data (train + test) and the forecasted values
plt.figure(figsize=(10, 6))

# Plot manual forecast
plt.plot(df['forecast_manual'], label='Manual Forecast', color='red')

# Plot actual values
df['actual_values'] = df['x']  # Assuming 'x' contains the actual target values
plt.plot(df['actual_values'], label='Actual Values', color='green')

# Plot confidence intervals
plt.fill_between(
    conf_int.index,
    conf_int.iloc[:, 0],  # Lower bound
    conf_int.iloc[:, 1],  # Upper bound
    color='pink',
    alpha=0.3,
    label='95% Confidence Interval'
)

# Add a legend, labels, and a title
plt.legend()
plt.title('ARIMA Forecast vs Actual with Confidence Interval')
plt.xlabel('Date')
plt.ylabel('Value')

# Show the plot
plt.show()
```

# Apply-Log-Square-Root-and-Cube-Root-Transformation

```python
df = pd.read_csv('../Stats_CA/Data/ts7.csv')
df.info()
del df['Unnamed: 0']

# apply Log Transformation and perform adf and packf check
#create log-transformed data
#https://www.statology.org/transform-data-in-python/
data_log = np.log(df)


#define grid of plots
fig, axs = plt.subplots(nrows=1, ncols=2)

#create histograms
axs[0].hist(df, edgecolor='black')
axs[1].hist(data_log, edgecolor='black')

#add title to each histogram
axs[0].set_title('Original Data')
axs[1].set_title('Log-Transformed Data')
```

```python
# Square Root Transformation

#create log-transformed data
data_qrt = np.sqrt(df)

#define grid of plots
fig, axs = plt.subplots(nrows=1, ncols=2)

#create histograms
axs[0].hist(df, edgecolor='black')
axs[1].hist(data_qrt, edgecolor='black')

#add title to each histogram
axs[0].set_title('Original Data')
axs[1].set_title('Square Root Transformed Data')
```

```python
# Cube Root Transformation
#create log-transformed data
data_cbrt = np.cbrt(df)

#define grid of plots
fig, axs = plt.subplots(nrows=1, ncols=2)

#create histograms
axs[0].hist(df, edgecolor='black')
axs[1].hist(data_cbrt, edgecolor='black')

#add title to each histogram
axs[0].set_title('Original Data')
axs[1].set_title('Cube Root Transformed Data')
```

```python
# perforem adf and pacf on log data - p value is greater than 0.05. data is still non-stationary 
test_adf(data_log['x']) 
```

```python
# https://www.geeksforgeeks.org/box-cox-transformation-using-python/
# 


 
# transform training data & save lambda value
fitted_data, fitted_lambda = stats.boxcox(df['x'])
 
# creating axes to draw plots
fig, ax = plt.subplots(1, 2)
 
# plotting the original data(non-normal) and 
# fitted data (normal)
sns.distplot(df['x'], hist = False, kde = True,
            kde_kws = {'shade': True, 'linewidth': 2}, 
            label = "Non-Normal", color ="green", ax = ax[0])
 
sns.distplot(fitted_data, hist = False, kde = True,
            kde_kws = {'shade': True, 'linewidth': 2}, 
            label = "Normal", color ="green", ax = ax[1])
 
# adding legends to the subplots
plt.legend(loc = "upper right")
 
# rescaling the subplots
fig.set_figheight(5)
fig.set_figwidth(10)
 
print(f"Lambda value used for Transformation: {fitted_lambda}")
```

# Perform-SARIMA

```python
# Split data train and test - last 30 intervals as test data and rest is training the model 
testtrain = (df.index < len(df)-30)
df_train = df[testtrain].copy()
df_test = df[~testtrain].copy()
```

```python
#https://medium.com/@poudel.birat25/three-techniques-to-improve-sarimax-model-for-time-series-forecasting-5d48db984fbe
# Fit SARIMA model

sarima_model = SARIMAX(df_train['x'],
                       order=(1, 1, 1),  
                       seasonal_order=(1, 1, 1, 30),  # daily seasonality (30 days)
                       enforce_stationarity=True,
                       enforce_invertibility=True)
sarima_result = sarima_model.fit()

# Print summary of the model
print(sarima_result.summary())

```

# 95-Confidence-Interval-for-SARIMA

```python
# Forecast on test data SARIMA
sarima_forecast_test = sarima_result.forecast(len(df_test))

df['sarima_forecast_manual'] = [None]*len(df_train) + list(sarima_forecast_test)

mae = mean_absolute_error(df_test, sarima_forecast_test)
mape = mean_absolute_percentage_error(df_test, sarima_forecast_test)
rmse = np.sqrt(mean_squared_error(df_test, sarima_forecast_test))

print(f'mae - manual: {mae}')
print(f'mape - manual: {mape}')
print(f'rmse - manual: {rmse}')
```

```python
# Generate forecast with confidence intervals using get_forecast()
forecast_obj = sarima_result.get_forecast(steps=len(df_test))  # Generate forecast object
forecast_test = forecast_obj.predicted_mean  # Extract forecasted values
conf_int = forecast_obj.conf_int(alpha=0.05)  # 95% confidence intervals

# Ensure forecast indices align with df_test
forecast_test.index = df_test.index
conf_int.index = df_test.index

# Add the forecasted values to the original DataFrame
df['sarima_forecast_manual'] = None  # Initialize the column
df.loc[df_test.index, 'sarima_forecast_manual'] = forecast_test  # Assign forecasted values

# Plotting the original data (train + test) and the forecasted values
plt.figure(figsize=(10, 6))

# Plot manual forecast
plt.plot(df['sarima_forecast_manual'], label='SARIMA Manual Forecast', color='red')

# Plot actual values
df['actual_values'] = df['x']  # Assuming 'x' contains the actual target values
plt.plot(df['actual_values'], label='Actual Values', color='green')

# Plot confidence intervals
plt.fill_between(
    conf_int.index,
    conf_int.iloc[:, 0],  # Lower bound
    conf_int.iloc[:, 1],  # Upper bound
    color='pink',
    alpha=0.3,
    label='95% Confidence Interval'
)

# Add a legend, labels, and a title
plt.legend()
plt.title('SARIMA Forecast vs Actual with Confidence Interval')
plt.xlabel('Date')
plt.ylabel('Value')

# Show the plot
plt.show()
```
